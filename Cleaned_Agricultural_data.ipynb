```python
import pandas as pd

# Load the dataset
file_path = 'Agricultural.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
# Get dataset information
print(df.info())

# Display summary statistics
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for seasonal procurement
plt.figure(figsize=(10, 6))
sns.barplot(x='Season', y='Amount(Rs)', data=df)  # Replace column names as per your dataset
plt.title('Procurement Quantity by Season')
plt.xlabel('Season')
plt.ylabel('Procurement Quantity')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line plot for procurement trends over time
plt.figure(figsize=(12, 8))
sns.lineplot(x='District', y='Commodity', data=df)  # Replace with actual column names
plt.title('Procurement Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Procurement Quantity')
plt.show()
```

```python
# Pie chart for procurement distribution by season
seasonal_data = df.groupby('Season')['Qty(MTs)'].sum()  # Replace column names accordingly
plt.figure(figsize=(8, 8))
seasonal_data.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Procurement by Season')
plt.ylabel('')
plt.show()
```
